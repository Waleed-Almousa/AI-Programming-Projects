```python
data = pd.read_html("https://climatology.nelson.wisc.edu/first-order-station-climate-data/madison-climate/lake-ice/history-of-ice-freezing-and-thawing-on-lake-mendota/")[0]
data=data[['Winter', 'Days of Ice Cover']]
data.dropna()
data.columns=['year', 'days']
data.dropna(inplace=True)
data['year'] = data['year'].apply(lambda x: int(x.split('-')[0]))
data.dropna()
# data['days'] = data['days'].astype(int)
data=data.iloc[1:]
data.sort_values(by="year", ascending=True, inplace=True)
```

```python
data.to_csv("hw5.csv", index=False)
```

```python
plt.plot(data['year'], data['days'])
plt.xlabel('Year')
plt.ylabel('Number of Frozen Days')
plt.savefig("plot.jpg")  # Save the plot as plot.jpg
```
